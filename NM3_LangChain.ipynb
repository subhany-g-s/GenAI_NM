```python
import requests
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Set your OpenRouter API key
OPENROUTER_API_KEY = "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"  # Replace with your actual API key
OPENROUTER_API_URL = "https://openrouter.ai/api/v1/chat/completions"
MODEL = "mistralai/mistral-7b-instruct:free"  # Change if needed (e.g., "gemini-pro", "claude-3-opus")

def generate_resume(job_title, skills, education, experience):
    """Generates a resume using OpenRouter's AI models."""

    prompt = f"""
    Create a professional resume with the following details:

    Job Title: {job_title}
    Skills: {", ".join(skills)}
    Education: {education}
    Work Experience: {experience if experience else "No experience"}

    Format it properly with sections: Summary, Skills, Experience, and Education.
    """

    headers = {
        "Authorization": f"Bearer {OPENROUTER_API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": MODEL,
        "messages": [{"role": "user", "content": prompt}]
    }

    response = requests.post(OPENROUTER_API_URL, json=payload, headers=headers)

    print("API Response Status Code:", response.status_code)  # Debugging
    print("API Response Text:", response.text)  # Debugging

    if response.status_code == 200:
        try:
            response_data = response.json()
            if "choices" in response_data and response_data["choices"]:
                return response_data["choices"][0]["message"]["content"]
            else:
                return "Error: No valid response received from OpenRouter."
        except requests.exceptions.JSONDecodeError:
            return "Error: Invalid JSON response from OpenRouter."
    else:
        return f"API Error: {response.status_code} - {response.text}"

def save_resume_as_pdf(resume_text, filename="resume.pdf"):
    """Saves the resume as a PDF file"""

    pdf = canvas.Canvas(filename, pagesize=letter)
    pdf.setFont("Helvetica", 12)

    y = 750  # Start position for text
    for line in resume_text.split("\n"):
        pdf.drawString(50, y, line)
        y -= 20  # Move to the next line

    pdf.save()
    print(f"Resume saved as {filename}")

# Example usage
if __name__ == "__main__":
    # Collect user input
    job_title = input("Enter job title: ")
    skills = input("Enter skills (comma separated): ").split(", ")
    education = input("Enter educational background: ")
    experience = input("Enter work experience (leave blank if none): ")

    # Generate and save the resume
    resume_text = generate_resume(job_title, skills, education, experience)
    print("\nGenerated Resume:\n")
    print(resume_text)

    if "Error" not in resume_text:
        save_resume_as_pdf(resume_text)
    else:
        print("Resume generation failed. Check API response.")
```

```python
import os
import requests
from langchain.chat_models import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage, SystemMessage

# Set OpenRouter API key
OPENROUTER_API_KEY = "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"  # Replace with your actual API key
os.environ["OPENAI_API_KEY"] = OPENROUTER_API_KEY  # LangChain expects an OpenAI-style key

# Initialize the LangChain chatbot with OpenRouter
chatbot = ChatOpenAI(
    model_name="mistralai/mistral-7b-instruct:free",  # You can also try "gemini-pro" or "claude-3-opus"
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key=OPENROUTER_API_KEY
)

# Chat history
chat_history = []

def chat_with_bot(user_input):
    """Handles conversation with OpenRouter via LangChain."""
    chat_history.append(HumanMessage(content=user_input))

    response = chatbot(chat_history)

    chat_history.append(AIMessage(content=response.content))

    return response.content

# Run chatbot loop
if __name__ == "__main__":
    print("ğŸ¤– Chatbot is ready! Type 'exit' to quit.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("ğŸ‘‹ Goodbye!")
            break

        response = chat_with_bot(user_input)
        print("Bot:", response)
```

```python
import os
import requests
from bs4 import BeautifulSoup
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# Set OpenRouter API key
OPENROUTER_API_KEY = "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"  # Replace with your actual API key
os.environ["OPENAI_API_KEY"] = OPENROUTER_API_KEY

# Choose a suitable model for summarization
MODEL_NAME = "mistralai/mistral-7b-instruct:free"  # You can also try "gemini-pro" or "claude-3-opus"

# Initialize LangChain AI model
summarizer = ChatOpenAI(
    model_name=MODEL_NAME,
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key=OPENROUTER_API_KEY
)

def fetch_article(url):
    """Fetches article content from a given URL."""
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        # Extract article text (modify if needed)
        paragraphs = soup.find_all("p")
        article_text = "\n".join([p.get_text() for p in paragraphs])

        return article_text.strip()
    except Exception as e:
        return f"Error fetching article: {e}"

def summarize_text(article_text):
    """Summarizes the given article using OpenRouter AI."""
    prompt = f"""
    Summarize the following news article in a concise and informative way:

    {article_text}

    The summary should be clear and easy to understand.
    """

    response = summarizer.invoke([HumanMessage(content=prompt)])
    return response.content

# Run summarization tool
if __name__ == "__main__":
    url = input("Enter the news article URL: ")

    # Fetch and summarize the article
    article_text = fetch_article(url)
    if "Error" in article_text:
        print(article_text)
    else:
        summary = summarize_text(article_text)
        print("\nğŸ“° News Summary:\n", summary)
```

```python
import os
import pdfplumber
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# Set OpenRouter API key
OPENROUTER_API_KEY = "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"  # Replace with your actual API key
os.environ["OPENAI_API_KEY"] = OPENROUTER_API_KEY

# Choose a suitable model for analysis
MODEL_NAME = "mistralai/mistral-7b-instruct:free"  # You can also try "gemini-pro" or "claude-3-opus"

# Initialize LangChain AI model
analyzer = ChatOpenAI(
    model_name=MODEL_NAME,
    openai_api_base="https://openrouter.ai/api/v1",
    openai_api_key=OPENROUTER_API_KEY
)

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF resume."""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            text = "\n".join([page.extract_text() for page in pdf.pages if page.extract_text()])
        return text.strip() if text else "No text found in the resume."
    except Exception as e:
        return f"Error extracting text: {e}"

def analyze_resume(resume_text):
    """Analyzes the resume using OpenRouter AI."""
    prompt = f"""
    Evaluate the following resume and provide feedback on strengths, weaknesses, and areas for improvement.

    Resume:
    {resume_text}

    Give suggestions to improve formatting, skills, and experience alignment for job applications.
    """

    response = analyzer.invoke([HumanMessage(content=prompt)])
    return response.content

# Run Resume Analyzer
if __name__ == "__main__":
    pdf_path = input("Enter the path to the resume PDF: ")

    resume_text = extract_text_from_pdf(pdf_path)
    if "Error" in resume_text:
        print(resume_text)
    else:
        feedback = analyze_resume(resume_text)
        print("\nğŸ“„ Resume Analysis:\n", feedback)
```
